## Notebook that summarizes applications

```python
import pandas as pd
import matplotlib.pyplot as plt
#load the Data
applicants = pd.read_csv(r"C:\Users\zohei\OneDrive\Documents\Data Wrangling\Lab 3\applicants.csv")
job_titles = pd.read_csv(r"C:\Users\zohei\OneDrive\Documents\Data Wrangling\Lab 3\job_titles.csv")
job_applications = pd.read_csv(r"C:\Users\zohei\OneDrive\Documents\Data Wrangling\Lab 3\applications.csv")

#print data
applicants.head()
job_titles.head()
job_applications.head()
```

```python
print("Merged Job Applications DataFrame:")
print(job_applications.head())
```

```python
# Count applications per job title using 'id_y' (from job_titles)
applications_per_job = job_applications['id_y'].value_counts().reset_index()
applications_per_job.columns = ['job_title_id', 'application_count']
```

```python
# Merge to get the title names (using id from job_titles)
applications_summary = applications_per_job.merge(job_titles, left_on='job_title_id', right_on='id', how='left')
```

```python
multiple_applications = applications_summary[applications_summary['application_count'] > 1]
print("Job Titles with More than One Application:")
print(multiple_applications[['job_applied_for', 'application_count']])
```

```python
plt.figure(figsize=(12, 6))
plt.bar(applications_summary['job_applied_for'], applications_summary['application_count'], color='skyblue')
plt.title('Number of Applications per Job Title')
plt.xlabel('Job Title')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(multiple_applications['application_count'], labels=multiple_applications['job_applied_for'], autopct='%1.1f%%', startangle=140)
plt.title('Job Titles with More than One Application')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show() #there are none in the dataset
```
